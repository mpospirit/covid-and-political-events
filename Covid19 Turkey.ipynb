```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"]=[8,4]
plt.rcParams["figure.dpi"]=300
```

```python
df=pd.read_csv("Resources/owid-covid-data.csv", parse_dates=True)
df["date"]=pd.to_datetime(df["date"])
df
```

```python
turkey=["Turkey"]
dfTurkey=df.loc[df["location"].isin(turkey)]
dfTurkey.head()
```

```python
dfTurkeyCases=dfTurkey.iloc[260:]
dfTurkeyCases.index = np.arange(0, len(dfTurkeyCases))
dfTurkeyCases.head()
```

```python
totalCases=dfTurkeyCases[["date","total_cases"]]
totalCases.set_index("date",inplace=True)
totalCases.plot(kind="area",legend=False)
plt.ticklabel_format(axis="y",style="plain")
plt.ylabel("Total Cases")
plt.xlabel("Date")
plt.grid(axis="y")
```

```python
dailyCases=dfTurkeyCases[["date","new_cases"]]
x=dailyCases["date"]
y=dailyCases["new_cases"]

fig,ax=plt.subplots()
ax.plot(x,y,color="0.7")

#record
recordDate,recordValue=dailyCases.max() #recordDate tablonun son tarihini veriyor neden anlamadım
recordIndex=dailyCases.loc[dailyCases["new_cases"]==recordValue].index[0]
ax.plot(dailyCases["date"][recordIndex],recordValue,"r.")
ax.text(dailyCases["date"][recordIndex],recordValue+1250,"rekor vaka\n({})".format(int(recordValue)), 
        fontsize=6, rotation=45, color="r")
#amiraller bildirisi
ax.text(dailyCases["date"][129],2000,"emekli amiraller bildirisi",fontsize=6,rotation="vertical",
        horizontalalignment="right",color="mediumaquamarine")
plt.axvline(x=dailyCases["date"][129],color="mediumaquamarine")
#quarantine
quarantine=dailyCases["date"][154]
plt.axvline(x=quarantine, color="purple")
ax.text(quarantine,2000,"tam kapanma başlangıcı",fontsize=6,rotation="vertical",horizontalalignment="right",color="purple")
#kongre
kongre=dailyCases["date"][118]
plt.axvline(x=kongre,color="y")
ax.plot(dailyCases["date"][133],dailyCases["new_cases"][133],"y.")
ax.text(kongre,2000,"akp kongresi",fontsize=6,rotation="vertical",horizontalalignment="right",color="y")
ax.text(dailyCases["date"][133],dailyCases["new_cases"][133]+1250,"kongre piki",fontsize=6,
        horizontalalignment="left",color="y",rotation=45)
#nevruz ve istanbul sözleşmesi
nevIst=dailyCases["date"][115]
plt.axvline(x=nevIst,color="g")
ax.plot(dailyCases["date"][128],dailyCases["new_cases"][128],"g.")
ax.text(nevIst,2000,"nevruz kutlamaları, istanbul sözleşmesi feshi protestoları",
        fontsize=6,rotation="vertical",horizontalalignment="right",color="g")
ax.text(dailyCases["date"][128],dailyCases["new_cases"][128]+1250,"nevruz,\nist. söz. piki",fontsize=6,
        horizontalalignment="right",color="g",rotation=-45)

plt.ylabel("günlük vaka")
plt.xlabel(" ")
ax.tick_params(axis="x", labelsize=7)
plt.ylim((0,75000))
plt.xlim(dailyCases["date"][50])
plt.grid(alpha=0.2)
ax.set_facecolor("1")
fig.patch.set_facecolor("0.8")
```
